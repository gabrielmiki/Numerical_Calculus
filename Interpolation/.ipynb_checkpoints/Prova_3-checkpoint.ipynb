    <h3>1. Teoria</h3>
</div>

<h4> <b> <em> 1.1 Espaços Vetoriais </em> </b> </h4>

<p> Um espaço vetorial (sobre o conjunto $ \mathbb{R} $ de escalares) é um conjunto $ V $ equipado com as operações de soma de vetores e de multiplicação por escalar e que satisfazem as propriedades usuais dos espaços $ \mathbb{R^{n}} $. </p>
<p> Dessa forma, um conjunto de funções de um intervalo $ I $ em $ \mathbb{R} $ também forma um espaço vetorial com as operações: </p>

<p> 
<ul> 
<li>
Dadas duas funções $ f $ e $ g $, a função $ f + g $ : $ I \rightarrow \mathbb{R} $ é definida por: 
<p style = "text-align: center;"> $ (f + g)(x) = f(x) + g(x) $ </p>
</li>
<li>
Dada uma $ f $ e um número real $ k $, a função $ k \cdot f $ : $ I \rightarrow \mathbb{R} $ é definida por: 
<p style = "text-align: center;" > $ (k \cdot f)(x) = k \cdot f(x) $ </p>
</li>
</ul>
</p>

<h4> <b> <em> 1.2 Bases </em> </b> </h4>

<p> Um conjunto $ B = (v_{1}, v_{2}, ... , v_{n}) \subset V $ é uma base no espaço vetorial se: </p>
<p> 
<ul>
<li>
$ B $ é LI
</li>
<li>
$ B $ gera $ V $
</li>
</ul>
</p>
<p> Para um espaço vetorial $ V $ formado por uma base $ B $ de $ n $ componentes, conclui-se que $ V $ possui dimensão igual a $ n $. </p> 

<h4> <b> <em> 1.3 Subespaços Vetoriais e Polinômios </em> </b> </h4>

<p> Consideramos $ V $ um espaço vetorial e $ H \subseteq V $ um subconjunto. Nós dizemos que $ H $ é uma subespaço vetorial de $ V $ se, além de ser um subconjunto, é por si só um espaço vetorial. </p>
<p> O conjunto $ \mathbb{P}(\mathbb{R}) $ das funções polinomiais com coeﬁcientes reais é um subespaço do espaço das funções contínuas. De fato, o polinômio nulo está neste espaço. Além disso, dados dois polinômios: </p>

<p style = "text-align: center;"> $ q(x) = a_{0} + a_{1}x + ... + a_{n} x^{n} $ e $ p(x) = b_{0} + b_{1}x + ... + b_{m} x^{m} $ </p> 

<p> Sabe-se que $ a \cdot q(x) + b \cdot p(x) $ vai ser um polinômio cujo grau é no máximo o maior dos valores $ m $ ou $ n $. </p> 

<h4> <b> <em> 1.4 Interpolação </em> </b> </h4>

<p> O problema algébrico da interpolação portanto, uma bijecção entre funções e vetores numéricos: </p>

<p style = "text-align: center;"> $ p \leftrightarrow \begin{bmatrix} w_{1} \\ w_{2} \\ ... \\ w_{n} \end{bmatrix}$ </p>

<p> Onde $ w_{1}, w_{2}, ..., w_{n} $ é definido por: <p>
    
<p style = "text-align: center;"> $ L_{i} \cdot p = w_{i} $ </p>

<p> Onde $ L_{i} $, com $ i = 1, ..., m, $ é o conjunto de funcionais lineares. </p>

<p> Isto é impossível se $ p $ pode ser qualquer função. Só é possível se $ p $ é restrita a pertence a um subespaço $ V $ de dimensão finita. </p>

<div style = "border: 1px solid black; padding: 10px; color: black; background-color: white; text-align: center; vertical-align:middle;">
<p> <em> <b> Teorema: </b> </em> O problema algébrico da interpolção tem solução para quaisquer $ w_{1}, w_{2}, ..., w_{n} $ se e só se a dimensão de $ V $ é igual ao número de funcionais lineares e o conjunto $ L_{i} $ é linearmente independente. Prova-se também que nesse caso a solução é única. </p> <div>

<div style = "border: 2px solid black; padding: 10px; color: black; background-color: white;text-align: center; vertical-align:middle;">
    <h3>2. Abordagem</h3>
</div>

<h4> <b> <em> 2.1 Calculo da Interpolada </em> </b> </h4>

<p> Para o calculo da interpolada $ p $, que assumimos como bem definida, utilizamos uma base de $ V $. Seja $ \phi_{1}, \phi_{2}, ..., \phi_{n} $ a base, a fim de se determinar $ p $, busca-se os coeficientes $ a_{0}, a_{1}, ..., a_{n} $ que proporcionam: </p>

<p style = "text-align: center;"> $ p(x) = \Sigma_{j = 0}^{n} a_{j} \phi_{j}(x) $

<h4> <b> <em> 2.2 Aplicação do Calculo da Interpolada no Problema Proposto </em> </b> </h4>

<p> Sabendo que a deflexão da viga é um polinômio quartico $ \mathbb{P}^{4} $, delimita-se o número de dimensões $ n $ como igual a $ 5 $, além da necessecidade de $ 5 $ funcionais, a fim de se obter uma única solução. </p>

<p> Dessa forma se estabelece: </p>

<p style = "text-align: center;"> $ y(x) \leftrightarrow \begin{bmatrix} y_{0} \\ d_{0} \\ y_{L} \\ d_{L} \\ y_{m} \end{bmatrix} $ </p>

<p> E definindo-se: </p>

<p style = "text-align: center;"> $ \phi_{1} = 1 $ </p>
<p style = "text-align: center;"> $ \phi_{2} = x $ </p>
<p style = "text-align: center;"> $ \phi_{3} = x^2 $ </p>
<p style = "text-align: center;"> $ \phi_{4} = x^3 $ </p>
<p style = "text-align: center;"> $ \phi_{5} = x^4 $ </p>

<p> Busca-se: </p>

<p style = "text-align: center;"> 
    $ \begin{bmatrix} L_{1}(\phi_{1}(x_{0})) & L_{1}(\phi_{2}(x_{0})) & L_{1}(\phi_{3}(x_{0})) & L_{1}(\phi_{4}(x_{0})) & L_{1}(\phi_{5}(x_{0})) \\ 
                      L_{2}(\phi_{1}(x_{1})) & L_{2}(\phi_{2}(x_{1})) & L_{2}(\phi_{3}(x_{1})) & L_{2}(\phi_{4}(x_{1})) & L_{2}(\phi_{5}(x_{1})) \\ 
                      L_{3}(\phi_{1}(x_{2})) & L_{3}(\phi_{2}(x_{2})) & L_{3}(\phi_{3}(x_{2})) & L_{3}(\phi_{4}(x_{2})) & L_{3}(\phi_{5}(x_{2})) \\
                      L_{4}(\phi_{1}(x_{3})) & L_{4}(\phi_{2}(x_{3})) & L_{4}(\phi_{3}(x_{3})) & L_{4}(\phi_{4}(x_{3})) & L_{4}(\phi_{5}(x_{3})) \\
                      L_{5}(\phi_{1}(x_{4})) & L_{5}(\phi_{2}(x_{4})) & L_{5}(\phi_{3}(x_{4})) & L_{5}(\phi_{4}(x_{4})) & L_{5}(\phi_{5}(x_{4})) \\ \end{bmatrix} \cdot 
      \begin{bmatrix} a_{0} \\ a_{1} \\ a_{2} \\ a_{3} \\ a_{4} \end{bmatrix} = 
      \begin{bmatrix} y_{0} \\ d_{0} \\ y_{L} \\ d_{L} \\ y_{m} \end{bmatrix} $ </p>

<p> Dessa forma, aplicando-se os funcionais ao problema proposto, obtém-se: </p>

<p  style = "text-align: center;">
    $ \begin{bmatrix} 1 & x_{0} & x_{0}^2 & x_{0}^3 & x_{0}^4 \\ 
                      0 & 1 & 2x_{1} & 3x_{1}^2 & 4x_{1}^3 \\ 
                      1 & x_{2} & x_{2}^2 & x_{2}^3 & x_{2}^4 \\
                      0 & 1 & 2x_{3} & 3x_{3}^2 & 4x_{3}^3 \\
                      0 & 0 & 2 & 6x_{4} & 12x_{4}^2 \\ \end{bmatrix} \cdot 
      \begin{bmatrix} a_{0} \\ a_{1} \\ a_{2} \\ a_{3} \\ a_{4} \end{bmatrix} = 
      \begin{bmatrix} y_{0} \\ d_{0} \\ y_{L} \\ d_{L} \\ y_{m} \end{bmatrix} $ </p>
      
<p> Logo em posse dos valores de $ x_{0}, x_{1}, x_{2}, x_{3}, x_{4} $ e $ y_{0}, d_{0}, y_{L}, d_{L}, y_{m} $ é possível econtrar os valores de $ a_{0}, a_{1}, a_{2}, a_{3}, a_{4} $, tal que: </p>

<p style = "text-align: center;"> $ y(x) = \Sigma_{j = 0}^{4} a_{j} \phi_{j}(x) $ </p>

    <h3>3. Código</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
def fazM(x):
    M = np.zeros((len(x), len(x)))
    
    for linha in range(len(x)):
        for coluna in range(len(x)):
            if linha == 0 or linha == 2 or linha == 4:
                M[linha, coluna] = x[linha] ** coluna
            
            elif linha == 1 or linha == 3:
                if coluna <= 1:
                    M[linha, coluna] = coluna
                else:
                    M[linha, coluna] = (x[linha] ** (coluna - 1)) * coluna 
    
    return M
```

```python
plt.figure(figsize = (20, 10))
plt.title("Assumindo um comprimento de 1.2 para a viga, tem-se M: ", fontweight = 800, fontsize = 17)
sns.heatmap(fazM([0, 0, 1.2, 1.2, 0.6]), annot = True)
```

```python
def plotInterpolada(num, x_in, a):
    x = np.linspace(0, 1.2, num)
    y = list()
    y_linha = list()
    y_2linha = list()
    y_in = list()
    
    for i in x:
        y.append((a[0]) + (a[1] * (i ** 1)) + (a[2] * (i ** 2)) + (a[3] * (i ** 3)) + (a[4] * (i ** 4)))
        y_linha.append((a[1]) + (2 * a[2] * i) + (3 * a[3] * (i ** 2)) + (4 * a[4] * (i ** 3)))
        y_2linha.append((2 * a[2]) + (6 * a[3] * i) + (12 * a[4] * (i ** 2)))
        
    for j in x_in:
        y_in.append((a[0] * j) + (a[1] * (j ** 1)) + (a[2] * (j ** 2)) + (a[3] * (j ** 3)) + (a[4] * (j ** 4)))
    
    plt.figure(figsize = (20, 10))
    
    plt.title(("Deflexão vertical encontrada para uma viga de comprimento igual a " + str(np.max(x))),
               fontweight = 800, fontsize = 17)
    
    plt.plot(x, y, color = 'lightgrey', linewidth = 3)
    plt.plot(x, y_linha, color = 'navy', linewidth = 3, alpha = 0.7)
    plt.plot(x, y_2linha, color = 'red', linewidth = 3, alpha = 0.5)
    plt.scatter(x_in, y_in, color = 'lightgrey', s = 130)
    
    return np.array(x), np.array(y), np.array(y_2linha)
```

```python
def encontraValoresMaxMin(data):
    y_min, x_y_min = data['y'].min(), float(data[data['y'] == data['y'].min()]['x'])
    y_max, x_y_max = data['y'].max(), float(data[data['y'] == data['y'].max()]['x'])
    
    y2linha_min, x_y2linha_min = data["y''"].min(), float(data[data["y''"] == data["y''"].min()]['x'])
    y2linha_max, x_y2linha_max = data["y''"].max(), float(data[data["y''"] == data["y''"].max()]['x'])
    
    print('O valor mínimo de y(x) é: ', y_min, '\nNo ponto: x = ', x_y_min,
          '\nJá o valor máximo de y(x) é: ', y_max, '\nNo ponto: x = ', x_y_max,
          '\n\nO valor mínimo de y2linha(x) é: ', y2linha_min, '\nNo ponto: x = ', x_y2linha_min,
          '\nJá o valor máximo de y2linha(x) é: ', y2linha_max, '\nNo ponto: x = ', x_y2linha_max)
    
    return y_min, y_max, y2linha_min, y2linha_max
```

```python
def resolveInterpolacao(x_func, num, y_funcionais):
    M = fazM(x_func)
    
    a = np.linalg.solve(M, y_funcionais)
    
    x, y, y_2linha = plotInterpolada(num, x_func, a)
    
    dados = np.vstack((x.T, np.vstack((y.T, y_2linha.T))))
    
    data = pd.DataFrame(data = dados.T, columns = ["x", "y", "y''"])
    
    y_min, y_max, y2linha_min, y2linha_max = encontraValoresMaxMin(data)
    
    return a, y_min, y_max, y2linha_min, y2linha_max
```

```python
_, _, _, _, _ = resolveInterpolacao([0, 0, 1.2, 1.2, 0.6], 100, [0, 0, 1, 0, 0.8])
```

    <h3>4. Calculo do Erro</h3>
</div>

<h4> <b> <em> 4.1 Abordagem </em> </b> </h4>

<p> A fim de se verificar o erro e variância do modelo optou-se por alterar o número de parâmetros calculados na interpolada e, com os resultados em mãos, averiguar mudaças quanto a locais de mínimo e máximo, tanto para a função em si, como para a curvatura da mesma. </p> 

```python
def analiseErroInterpolada(x_func, y_funcionais, num_pts):
    pts = list()
    y_min_list = list() 
    y_max_list = list() 
    y2linha_min_list = list() 
    y2linha_max_list = list()
    
    num = 0
    while num <= num_pts:
        num += 1
        
        pts.append(10 * num)
        print('\nPara um total de ', 10 * num, ' pontos:')
        _, y_min, y_max, y2linha_min, y2linha_max = resolveInterpolacao(x_func, 10 * num, y_funcionais)
        
        y_min_list.append(y_min)
        y_max_list.append(y_max) 
        y2linha_min_list.append(y2linha_min) 
        y2linha_max_list.append(y2linha_max)
        
    y_min_list = np.array(y_min_list)
    y_max_list = np.array(y_max_list)
    y2linha_min_list = np.array(y2linha_min_list)
    y2linha_max_list = np.array(y2linha_max_list)
    
    dados = np.vstack((np.vstack((y_min_list.T, y_max_list.T)), np.vstack((y2linha_min_list.T, y2linha_max_list.T))))
    data = pd.DataFrame(data = dados.T, columns = ["y_min", "y_max", "y2linha_min", "y2linha_max"])
    data.index = pts
    
    return data
```

```python
results_finais = analiseErroInterpolada([0, 0, 1.2, 1.2, 0.6], [0, 0, 1, 0, 0.8], 20)
```

<h4> <b> <em> 4.2 Discussão dos Resultados </em> </b> </h4>

<p> Com os valores de mínimo e máximo, tanto para a função, como para a segunda derivada da mesma, analisou-se a variação para cada resultado de acordo com o número de pontos utilizados dentro do intervalo de comprimento da barra. </p> 

```python
plt.figure(figsize = (20, 10))
    
plt.title("Variação dos valores de mínimo encontrados, de acordo com o número de pontos no intervalo L",
               fontweight = 800, fontsize = 17)

    
plt.plot(results_finais.index, 
         results_finais["y_min"].values, color = 'lightgrey', linewidth = 3)
```

```python
plt.figure(figsize = (20, 10))
    
plt.title("Variação dos valores de máximos encontrados, de acordo com o número de pontos no intervalo L",
               fontweight = 800, fontsize = 17)

    
plt.plot(results_finais.index, 
         results_finais["y_max"].values, color = 'lightgrey', linewidth = 3)
```

```python
plt.figure(figsize = (20, 10))
    
plt.title("Variação dos valores de máximos encontrados, de acordo com o número de pontos no intervalo L",
               fontweight = 800, fontsize = 17)

    
plt.plot(results_finais.index, 
         results_finais["y2linha_min"].values, color = 'lightgrey', linewidth = 3)
```

```python
plt.figure(figsize = (20, 10))
    
plt.title("Variação dos valores de máximos encontrados, de acordo com o número de pontos no intervalo L",
               fontweight = 800, fontsize = 17)

    
plt.plot(results_finais.index, 
         results_finais["y2linha_max"].values, color = 'lightgrey', linewidth = 3)
```

<h4> <b> <em> 4.3 Conclusão </em> </b> </h4>

<p> Como observado alguns dos valores extremos calculados variam de acordo com o número de pontos utilizados dentro do intervalo do comprimento da barra. Contudo, as diferenças são pequenas e convergem rapidamente para um valor fixo, como visto nos gráficos acima, o resultado obtido é quase constante a partir de cinquenta partições do comprimento original.  </p>
