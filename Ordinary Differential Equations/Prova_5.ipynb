<h4> <b> <em> 1.1 Equações Diferenciais Ordinárias </em> </b> </h4>

<p> Equações envolvendo derivadas de uma função incógnita, nelas todas as derivadas são relativas a uma única variável independente, por vezes representando o tempo. Solução numérica de equações diferencias é baseada numa aproximação de dimensão finita em que equação diferencial é substituída por uma equação algébrica cuja solução aproxima a solução da equação diferencial. </p> 

<p> Ordem de uma EDO é determinada pela ordem da mais alta derivada da função solução que ocorre na EDO. Para uma EDO de ordem k: </p> 

<p style = "text-align: center"> $ y^k (t) = f(1, y, y', ..., y^{(k - 1)}) $ </p>

<p> Define-se k novas funções incógnitas: </p>

<p style = "text-align: center"> $ u_{1} = y(t), u_{2} = y'(t), ..., u_{k} = y^{(k - 1)}(t)) $ </p>

<p> A EDO original é então equivalente ao sistema de primeira ordem. </p>

<p style = "text-align: center"> $$ \begin{bmatrix} u_{1}' \\ u_{2}' \\ ... \\ u_{k - 1}' \end{bmatrix} = 
                                    \begin{bmatrix} u_{2} \\ u_{3} \\ ... \\ u_{k} \end{bmatrix} $$ </p>
                                    
<p> Por si só a EDO $ y' = f(t,y) $ não determina uma função solução única. Isto porque a EDO apenas especifica o declive $ y'(t) $ da função solução em cada ponto, mas não especifica o valor de $ y(t) $ para algum ponto. Para obter uma solução particular, o valor $ y_{0} $ da função solução tem de ser conhecido para algum ponto $ t_{0} $. </p>

<h4> <b> <em> 1.2 Problema de Valor Inicial </em> </b> </h4>

<p> Se considerarmos a variável independente $ t $ como o tempo, podemos pensar em $ t_{0} $ como o tempo inicial e em $ y_{0} $ como o valor inicial da função incógnita, por isso, é designado por Problema de Valor Inicial. </p>

<p> A EDO governa a evolução do sistema ao longo do tempo desde o seu estado inicial $ y_{0} $ no tempo $ t_{0} $. </p>

<h4> <b> <em> 1.3 Soluções Numéricas </em> </b> </h4>

<p> Solução analítica de EDO é uma função bem definida que pode ser avaliada para qualquer valor de $ t $. Já a solução numérica de EDO é uma tabela de valores aproximados da função solução para num conjunto discretos de pontos. </p>

<p> Calculo de $ f(t_{0}, y_{0}) $ indica o declive da trajectória nesse ponto. Usa-se desta informação para prever o valor $ y1 $ da solução no tempo futuro $ t_{1} = t_{0} + h $ para um determinado incremento de tempo $ h $. </p>

<h4> <b> <em> 1.4 Método de Euler </em> </b> </h4>

<p> Método de Euler prevê solução através da extrapolação ao longo de uma linha reta cujo declive é $ f(t_{k} , y_{k}) $. </p>

<p style = "text-align: center"> $ y_{k + 1} = y_{k} + h_{k} f(t_{k}, y_{k}) $ </p>

<h5> <b> <em> 1.4.1 Método de Euler Explícito </em> </b> </h5>

<p> Método de Euler é explícito porque apenas usa a informação no tempo $ t_{k} $ para avançar a solução no tempo $ t_{k+1} $. Devido a isso o método de Euler apresenta uma região de estabilidade muito limitada. </p>

<h5> <b> <em> 1.4.2 Método de Euler Implícito </em> </b> </h5>

<p> Maiores regiões de estabilidade podem ser obtidas usando informação do $ t_{k+1} $, o que torna o método implícito. O exemplo mais simples é o método de Euler implícito: </p>

<p style = "text-align: center"> $ y_{k + 1} = y_{k} + h_{k} f(t_{k + 1}, y_{k + 1}) $ </p>

<p> O método é implícito porque necessitamos de avaliar $ f $ com argumento $ y_{k+1} $ antes de conhecer o seu valor. </p>

<h4> <b> <em> 1.5 Erros na Solução Numérica </em> </b> </h4>

<p> Métodos numéricos para resolver EDO's incorrem em dois tipos de erros distintos, Erros de arredondamento (devidos à precisão finita da aritmética de ponto flutuante) e Erros de truncatura (devidos aos métodos de aproximação usados e que permaneceriam mesmo que se usasse uma aritmética exata). </p>

<h5> <b> <em> 1.5.1 Erro de Truncatura </em> </b> </h5>

<p> Erro de truncatura em qualquer ponto $ t_{k} $ pode ser decomposto em: </p>

<ul> 
    <li> Erro Global: diferença entre solução calculada $ y_{k} $ e solução verdadeira $ y(t_{k}) $ que passa pelo ponto $ (t_{0},y_{0}) $. </li> 
</ul>

<p style = "text-align: center"> $ e_{k} = y_{k} - y(t_{k}) $ </p>
    
<ul> 
    <li> Erro Local: erro efectuado num único passo do método numérico. </li>
</ul>

<p style = "text-align: center"> $ l_{k} = y_{k} - u_{k - 1}(t_{k}) $ </p>

<h4> <b> <em> 1.6 Modelo SEIR </em> </b> </h4>

<p> O modelo SEIR busca mapear doenças que têm o que é chamado de uma fase latente ou exposta, durante o qual é dito ao indivíduo estar infetado mas não infecioso. Neste modelo, a população anfitriã ($ N $) é dividida em quatro compartimentos: suscetíveis, expostos, infectante e recuperados, com o número de indivíduos num compartimento, ou as suas densidades denotados, respectivamente, por $ S(t) $, $ E(t) $, $ I(t) $, $ R(t) $, em que $ N = S(t) + E(t) + I(t) + R(t) $. As definições de cada grupo são: </p>

<ul> 
<li> <b> S(t) </b> é usado para representar o número de indivíduos não infectados com a doença no momento t, ou aqueles suscetíveis à doença. </li>
    
<li> <b> E(t) </b> representa o número de indivíduos que tenham sido infectadas com a doença. </li>

<li> <b> I(t) </b> representa o número de indivíduos que podem transmitir a doença. </li>   
    
<li> <b> R(t) </b> é o compartimento utilizado para aqueles indivíduos que foram infectados e, de seguida, removidos a partir da doença, quer devido à imunização ou devido à morte. Os que estão nesta categoria não são capazes de ser infectados novamente ou para transmitir a infecção a outras pessoas. </li>      
</ul>

<p> Em tal modelo a variação dos grupos é dada por: </p>

<p style = "text-align: center"> $ \frac{dS}{dT} = \frac{- \beta S I}{N} $ </p>
<p style = "text-align: center"> $ \frac{dE}{dT} = \frac{- \beta S I}{N} - a E $ </p>
<p style = "text-align: center"> $ \frac{dI}{dT} = a E - \gamma I $ </p>
<p style = "text-align: center"> $ \frac{dR}{dT} = \gamma I $ </p>

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
class modeloSEIR():
    
# =========================================================================================================
#                                          Configuração do Modelo
# =========================================================================================================

    def __init__(self, betha, N, a, gamma, deltaT, totalT, Si, Ei, Ii, Ri):
        
        # ===================================================================
        # Vetor para armazenar variações do grupo Sucetível. Valor inicial
        # ===================================================================
        
        self._susceptibleEulerGroup = list()
        self._susceptibleRK2Group = list()
        self._susceptibleGroupValue = Si
        
        # ===================================================================
        # Vetor para armazenar variações do grupo Infectado. Valor inicial 
        # ===================================================================
        
        self._exposedEulerGroup = list()
        self._exposedRK2Group = list()
        self._exposedGroupValue = Ei
        
        # ===================================================================
        # Vetor para armazenar variações do grupo Transmitindo. Valor inicial
        # ===================================================================
        
        self._infectiousEulerGroup = list()
        self._infectiousRK2Group = list()
        self._infectiousGroupValue = Ii
        
        # ===================================================================
        # Vetor para armazenar variações do grupo Recuperado. Valor inicial
        # ===================================================================
        
        self._recoveredEulerGroup = list()
        self._recoveredRK2Group = list()
        self._recoveredGroupValue = Ri
        
        # ===================================================================
        # Parâmetro do modelo
        # ===================================================================        
        
        self._betha = betha
        self._N = N
        self._a = a
        self._gamma = gamma
        self._deltaT = deltaT
        self._totalT = totalT
        
    def reset(self, betha, N, a, gamma, deltaT, totalT, Si, Ei, Ii, Ri):
        
        self._susceptibleEulerGroup = list()
        self._susceptibleRK2Group = list()
        self._susceptibleGroupValue = Si
        
        self._exposedEulerGroup = list()
        self._exposedRK2Group = list()
        self._exposedGroupValue = Ei
        
        self._infectiousEulerGroup = list()
        self._infectiousRK2Group = list()
        self._infectiousGroupValue = Ii
        
        self._recoveredEulerGroup = list()
        self._recoveredRK2Group = list()
        self._recoveredGroupValue = Ri
        
        self._betha = betha
        self._N = N
        self._a = a
        self._gamma = gamma
        self._deltaT = deltaT
        self._totalT = totalT

# =========================================================================================================
#                                          Método de Euler Explícito
# =========================================================================================================

    def susceptibleEulerExplicitoVariation(self, S, I):   
        Sf = S - ((self._deltaT * self._betha * I * S) / self._N)

        self._susceptibleEulerGroup.append(Sf)
        
        return Sf
    
    def exposedEulerExplicitoVariation(self, E, I, S):    
        Ef = E + ((self._deltaT * self._betha * I * S) / self._N) - (self._deltaT * self._a * E)
        
        self._exposedEulerGroup.append(Ef)
        
        return Ef
    
    def infectiousEulerExplicitoVariation(self, I, E):
        If = I + (self._deltaT * self._a * E) - (self._deltaT * self._gamma * I)

        self._infectiousEulerGroup.append(If)
        
        return If
        
    def recoveredEulerExplicitoVariation(self, R, I):
        Rf = R + (self._deltaT * self._gamma * I)
        
        self._recoveredEulerGroup.append(Rf)
        
        return Rf
    
# =========================================================================================================
#                                          Método RK2
# =========================================================================================================

    def susceptibleRK2Variation(self, S, S1, S2):   
        deltaS1 = S1 - S
        deltaS2 = S2 - S1
        deltaTotal = deltaS1 + deltaS2
        
        Sf = S + (self._deltaT * ((deltaTotal) / 2))

        self._susceptibleRK2Group.append(Sf)
        
        return Sf
    
    def exposedRK2Variation(self, E, E1, E2):    
        deltaE1 = E1 - E
        deltaE2 = E2 - E1
        deltaTotal = deltaE1 + deltaE2
        
        Ef = E + (self._deltaT * ((deltaTotal) / 2))
        
        self._exposedRK2Group.append(Ef)
        
        return Ef
    
    def infectiousRK2Variation(self, I, I1, I2):
        deltaI1 = I1 - I
        deltaI2 = I2 - I1
        deltaTotal = deltaI1 + deltaI2
        
        If = I + (self._deltaT * ((deltaTotal) / 2))

        self._infectiousRK2Group.append(If)
        
        return If
        
    def recoveredRK2Variation(self, R, R1, R2):
        deltaR1 = R1 - R
        deltaR2 = R2 - R1
        deltaTotal = deltaR1 + deltaR2
        
        Rf = R + (self._deltaT * ((deltaTotal) / 2))
        
        self._recoveredRK2Group.append(Rf)
        
        return Rf
    
# =========================================================================================================
#                                      Aplicação dos Métodos no Modelo
# =========================================================================================================

    def mainEulerExplicito(self):
        for t in range(self._totalT):
            S = self.susceptibleEulerExplicitoVariation(self._susceptibleGroupValue, self._infectiousGroupValue)
            E = self.exposedEulerExplicitoVariation(self._exposedGroupValue, self._infectiousGroupValue, self._susceptibleGroupValue)
            I = self.infectiousEulerExplicitoVariation(self._infectiousGroupValue, self._exposedGroupValue)
            R = self.recoveredEulerExplicitoVariation(self._recoveredGroupValue, self._infectiousGroupValue)
            
            self._susceptibleGroupValue = S
            self._exposedGroupValue = E
            self._infectiousGroupValue = I
            self._recoveredGroupValue = R
            
    def mainRK2(self):
        for t in range(self._totalT):
            S1 = self.susceptibleEulerExplicitoVariation(self._susceptibleGroupValue, self._infectiousGroupValue)
            E1 = self.exposedEulerExplicitoVariation(self._exposedGroupValue, self._infectiousGroupValue, self._susceptibleGroupValue)
            I1 = self.infectiousEulerExplicitoVariation(self._infectiousGroupValue, self._exposedGroupValue)
            R1 = self.recoveredEulerExplicitoVariation(self._recoveredGroupValue, self._infectiousGroupValue)
            
            S2 = self.susceptibleEulerExplicitoVariation(S1, I1)
            E2 = self.exposedEulerExplicitoVariation(E1, I1, S1)
            I2 = self.infectiousEulerExplicitoVariation(I1, E1)
            R2 = self.recoveredEulerExplicitoVariation(R1, I1)
                        
            self._susceptibleGroupValue = self.susceptibleRK2Variation(self._susceptibleGroupValue, S1, S2)
            self._exposedGroupValue = self.exposedRK2Variation(self._exposedGroupValue, E1, E2)
            self._infectiousGroupValue = self.infectiousRK2Variation(self._infectiousGroupValue, I1, I2)
            self._recoveredGroupValue = self.recoveredRK2Variation(self._recoveredGroupValue, R1, R2)
            
            
# =========================================================================================================
#                                      Análise de Resultados
# =========================================================================================================

    def achaDiferençasEuler(self, group1, group2, intervalo):
        for t in range(len(group1)):
            if (abs(group1[t] - group2[t]) > 
                (intervalo * group1[t])):
        
                print("\n==========================================================================================================\n",
                      "A diferença entre os métodos é maior do que ", 100 * intervalo, "% do valor encontrado pelo método de Euler em t =", t,
                      "\n==========================================================================================================\n")
        
                break
            
    def achaDiferençasRK2(self, group1, group2, intervalo):
        for t in range(len(group1)):
            if (abs(group1[t] - group2[t]) > 
                (intervalo * group1[t])):
        
                print("\n==========================================================================================================\n",
                      "A diferença entre os métodos é maior do que ", 100 * intervalo, "% do valor encontrado pelo método de RK2 em t =", t,
                      "\n==========================================================================================================\n")
        
                break
           
    def achaEquilibrioSus(self, group, intervalo):
        maxVal = max(group)
        minVal = min(group)

        for t in range(len(group)):
            if (group[t] >= ((maxVal + minVal) / 2)):
                continue
            
            else:
                if ((group[t] <= (group[t - 1] + (intervalo * group[t - 1]))) and 
                    (group[t] >= (group[t - 1] - (intervalo * group[t - 1])))):
                    
                    return (t - 1), group[t - 1]
    
    def achaEquilibrioRec(self, group, intervalo):
        maxVal = max(group)
        minVal = min(group)

        for t in range(len(group)):
            if (group[t] <= ((maxVal + minVal) / 2)):
                continue
            
            else:
                if ((group[t] <= (group[t - 1] + (intervalo * group[t - 1]))) and 
                    (group[t] >= (group[t - 1] - (intervalo * group[t - 1])))):
                    
                    return (t - 1), group[t - 1]
                
    def achaEquilibrioConcavo(self, group, intervalo):
        maxVal = max(group)
        maxValIndex = group.index(max(group))
        minVal = min(group)
        
        for t in range(len(group)):
            if (t < maxValIndex):
                continue
            
            else:                
                if (group[t] >= ((maxVal + minVal) / 2)):
                    continue
                    
                else:  
                    if ((group[t] <= (group[t - 1] + (intervalo * group[t - 1]))) and 
                        (group[t] >= (group[t - 1] - (intervalo * group[t - 1])))):
                        
                        return (t - 1), group[t - 1]
                
    def plotEulerResults(self, title, intervalo, details = False):
        plt.figure(figsize = (20, 10))
        
        plt.title(title, fontweight = 800, fontsize = 17)
        
        plt.plot(self._susceptibleEulerGroup, color = "red", linewidth = 2, alpha = 0.4, label = ("Susceptible Group"))
        plt.plot(self._exposedEulerGroup, color = "lightgrey", linewidth = 2, label = ("Exposed Group"))
        plt.plot(self._infectiousEulerGroup, color = "navy", linewidth = 2, alpha = 0.4, label = ("Infectious Group"))
        plt.plot(self._recoveredEulerGroup, color = "green", linewidth = 2, alpha = 0.4, label = ("Recovered Group"))
        
        if (details):
            maxExposed = max(self._exposedEulerGroup)
            maxTExposed = self._exposedEulerGroup.index(maxExposed)
        
            maxInfectious = max(self._infectiousEulerGroup)
            maxTInfectious = self._infectiousEulerGroup.index(maxInfectious)
       
            equiTempSus, equiValSus = self.achaEquilibrioSus(self._susceptibleEulerGroup, intervalo)
            equiTempRec, equiValRec = self.achaEquilibrioRec(self._recoveredEulerGroup, intervalo)        

            xSusceptible = equiTempSus * np.ones(100)
            ySusceptible = np.linspace(0, equiValSus, 100)
        
            xRecovered = equiTempRec * np.ones(100)
            yRecovered = np.linspace(0, equiValRec, 100)
        
            plt.plot(xSusceptible, ySusceptible, color = "red", linestyle = "--", linewidth = 4, alpha = 0.2)
            plt.scatter(equiTempSus, equiValSus, color = 'red', s = 40, alpha = 0.2)
        
            plt.plot(xRecovered, yRecovered, color = "green", linestyle = "--", linewidth = 4, alpha = 0.2)
            plt.scatter(equiTempRec, equiValRec, color = 'green', s = 40, alpha = 0.2)
            
            plt.scatter(maxTExposed, maxExposed, color = 'lightgrey', s = 40)
            plt.scatter(maxTInfectious, maxInfectious, color = 'navy', s = 40)
        
        plt.legend()
        plt.show()

    def plotRK2Results(self, title, intervalo, details = False):
        plt.figure(figsize = (20, 10))
        
        plt.title(title, fontweight = 800, fontsize = 17)
        
        plt.plot(self._susceptibleRK2Group, color = "red", linewidth = 2, alpha = 0.4, label = ("Susceptible Group"))
        plt.plot(self._exposedRK2Group, color = "lightgrey", linewidth = 2, label = ("Exposed Group"))
        plt.plot(self._infectiousRK2Group, color = "navy", linewidth = 2, alpha = 0.4, label = ("Infectious Group"))
        plt.plot(self._recoveredRK2Group, color = "green", linewidth = 2, alpha = 0.4, label = ("Recovered Group"))
        
        if (details):
            maxExposed = max(self._exposedRK2Group)
            maxTExposed = self._exposedRK2Group.index(maxExposed)
        
            maxInfectious = max(self._infectiousRK2Group)
            maxTInfectious = self._infectiousRK2Group.index(maxInfectious)
        
            equiTempSus, equiValSus = self.achaEquilibrioSus(self._susceptibleRK2Group, intervalo)    
            equiTempRec, equiValRec = self.achaEquilibrioRec(self._recoveredRK2Group, intervalo)        

            xSusceptible = equiTempSus * np.ones(100)
            ySusceptible = np.linspace(0, equiValSus, 100)
        
            xRecovered = equiTempRec * np.ones(100)
            yRecovered = np.linspace(0, equiValRec, 100)
        
            plt.plot(xSusceptible, ySusceptible, color = "red", linestyle = "--", linewidth = 4, alpha = 0.2)
            plt.scatter(equiTempSus, equiValSus, color = 'red', s = 40, alpha = 0.2)
        
            plt.plot(xRecovered, yRecovered, color = "green", linestyle = "--", linewidth = 4, alpha = 0.2)
            plt.scatter(equiTempRec, equiValRec, color = 'green', s = 40, alpha = 0.2)
       
            plt.scatter(maxTExposed, maxExposed, color = 'lightgrey', s = 40)
            plt.scatter(maxTInfectious, maxInfectious, color = 'navy', s = 40)
        
        plt.legend()
        plt.show()

```
